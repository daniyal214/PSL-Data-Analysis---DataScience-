```python
dic = {}
for i in teams:
    ans = psl[(psl['innings1'] == i) | (psl['innings2'] == i)]['result'].count()
    won = psl[psl['winner'] == i]['result'].count()
    percent = (won/ans * 100).round(2)
    batting = psl[psl['winner'] == i]['Win-Batted First'].count()
    per_bat = (batting/won * 100).round(2)
    bowling = psl[psl['winner'] == i]['Win-Bowled First'].count()
    per_bowl = (bowling/won * 100).round(2)
    print(f'{i} played {ans} matches and won {won} matches -------> {percent}%')
    print(f'Out of {won} won matches {batting} won on batting first ----------> {per_bat} ')
    print(f'Out of {won} won matches {bowling} won on bowling first ----------> {per_bowl}\n')
    
    if i not in dic:
        dic[i] = {'Matches': ans, 'Won':won, 'Bat': batting, 'Bowl':bowling, 'Win%':percent, 'Win-Bat%':per_bat,\
                 'Win-Bowl%':per_bowl}
```
